```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# üíÖ Settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)
```

```python
# üìÅ Load data
df = pd.read_csv("../data/AAPL.csv", parse_dates=["Date"])
df.set_index("Date", inplace=True)
df = df.sort_index()
```

```python
# üîç Basic Checks
print("üîπ Head of Data:")
display(df.head())

print("üîπ Info:")
df.info()

print("üîπ Summary Stats:")
display(df.describe())

print("üîπ Missing Values:")
print(df.isna().sum())

print("üîπ Is date sorted:", df.index.is_monotonic_increasing)
```

```python
# üìà Close Price Plot
plt.plot(df["Close"], label="Close Price")
plt.title("AAPL - Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# üìâ Volume Trend
plt.plot(df["Volume"], label="Volume", color="orange")
plt.title("AAPL - Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# ‚ö†Ô∏è Missing Dates Check
expected_days = pd.date_range(start=df.index.min(), end=df.index.max(), freq="B")
missing_days = expected_days.difference(df.index)
print(f"üîé Missing trading days: {len(missing_days)}")
if len(missing_days):
    print(missing_days)
```
